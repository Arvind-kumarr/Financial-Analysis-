```python
# will replace this 123 with caragorical values
data['Education'].replace({1:'Undergraduate',2:'graduate',3:'professional'},inplace=True)
data['Education'].value_counts()
```

```python
#now we focus on the another coloumns which are securities and cd account

data.columns
#for that we define the function over here

def security(y):
  if( y['Securities Account']==1 and y['CD Account'] ==1):
    return"Hold Securities and deposits"
  elif(y['Securities Account']==0 and y['CD Account'] ==0):
    return "Does not hold securities and deposits"
  elif(y['Securities Account']==1 and y['CD Account'] ==0):
    return"only hold Securities "
  elif(y['Securities Account']==0 and y['CD Account'] ==1):
    return "only holds the deposit"

```

```python
# now apply this function in your dataFrame
data['Account_holder_category']= data.apply(security, axis=1)
```

```python
# for each column we cant be like to write multiple time the same function so we will define our function
def plot(col1,col2,label1,label2,title):
  sns.distplot(data[data[col2]==0][col1], hist=False , label = label1)
  sns.distplot(data[data[col2]==1][col1], hist=False , label = label2 )
  plt.legend()
  plt.title(title)
```
